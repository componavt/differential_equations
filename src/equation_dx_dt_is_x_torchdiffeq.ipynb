```python
# Установка torchdiffeq
!pip install git+https://github.com/rtqichen/torchdiffeq.git

import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torchdiffeq import odeint

# Определение функции правой части ОДУ
class ODEFunc(nn.Module):
    def forward(self, t, x):
        return x

# Начальное условие
x0 = torch.tensor([1.0])

# Интервал интегрирования
t = torch.linspace(0, 3, 300)

# Решение ОДУ
func = ODEFunc()
x = odeint(func, x0, t)

# Точное решение
x_exact = torch.exp(t)

# Визуализация
plt.figure(figsize=(12, 5))
plt.plot(t.numpy(), x_exact.numpy(), label='Точное решение $e^t$', linewidth=2)
plt.plot(t.numpy(), x.numpy(), '--', label='Neural ODE решение', linewidth=2)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Решение ОДУ dx/dt = x с помощью torchdiffeq')
plt.legend()
plt.grid(True)
plt.show()
```
