```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torchdiffeq import odeint
import numpy as np

# Параметры решения
t_start = 0
t_end = 10
steps = 1000

# Одно начальное условие — x(0) = 1.0
x0 = torch.tensor([1.0], dtype=torch.float32)

# Время в float32
t = torch.linspace(t_start, t_end, steps, dtype=torch.float32)

# ОДУ: dx/dt = sin(x)
class ODEFunc(nn.Module):
    def forward(self, t, x):
        print(f"t: {t}, x: {x}, dx/dt: {torch.sin(x)}")
        return torch.sin(x)


# Решение
func = ODEFunc()
x_torch = odeint(func, x0, t)

# Визуализация
plt.figure(figsize=(12, 6))
plt.plot(t.numpy(), x_torch.numpy(), label=r'$x(0) = 1.0$', linewidth=2)
plt.title('Решение ОДУ $\\frac{dx}{dt} = \sin(x)$')
plt.xlabel('Время t')
plt.ylabel('x(t)')
plt.grid(True)
plt.legend()
plt.show()
```
